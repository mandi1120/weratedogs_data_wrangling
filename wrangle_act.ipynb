## Analyzing and Visualizing Data
In this section, analyze and visualize your wrangled data. You must produce at least **three (3) insights and one (1) visualization.**

### Insights:
1. More tweets are done earlier in the week than later in the week. Mondays had the most tweets at 376.

### Visualization

```python
import matplotlib.pyplot as plt

# identify the tweets day of week and date
master_df['tweet_day_of_wk_num'] = master_df["timestamp_new"].dt.dayofweek
master_df['tweet_day_of_wk'] = master_df['tweet_day_of_wk_num'].map({
    0: '1 - Monday',
    1: '2 - Tuesday',
    2: '3 - Wednesday',
    3: '4 - Thursday',
    4: '5 - Friday',
    5: '6 - Saturday',
    6: '7 - Sunday'
})

master_df['tweet_month_yr'] = master_df['timestamp_new'].dt.to_period('M')
master_df.head()
```

```python
# Tweets by Day of Week
print(master_df.groupby(['tweet_day_of_wk']).size())
chart = master_df['tweet_day_of_wk'].value_counts().sort_index(ascending=False)
chart.plot(kind='barh', title='Tweets by Day of Week', color='navy')
plt.show()
```

```python
# Tweets by Dog Stage
print(master_df.groupby(['dog_stage']).size())
chart = master_df['dog_stage'].value_counts().sort_index(ascending=False)
chart.plot(kind='bar', title='Tweets by Dog Stage', color='navy')
plt.show()
```

```python
# Tweet Trend by Month
print(master_df.groupby(['tweet_month_yr']).size())
plt.figure(figsize=(12,6))
chart = master_df['tweet_month_yr'].value_counts().sort_index(ascending=True)
chart.plot(kind='line', title='Tweet Trend by Month', color='navy')
plt.show()
```
